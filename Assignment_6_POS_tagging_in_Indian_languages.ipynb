```python
import nltk
nltk.download('indian')
from nltk.corpus import indian
from nltk.tag import tnt
import string
```

# **Step 2.1 : Importing dataset for POS Tagging**
***Language:*** Hindi

```python
tagged_set = '/content/hindi.pos'
word_set = indian.sents(tagged_set)
count = 0
for sen in word_set:
    count = count + 1
    sen = "".join([" "+i if not i.startswith("'") and i not in string.punctuation else i for i in sen]).strip()
    print (count, sen)
print ('Total sentences in the tagged file are',count)

train_perc = .9

train_rows = int(train_perc*count)
test_rows = train_rows + 1

print ('Sentences to be trained',train_rows, 'Sentences to be tested against',test_rows)
```

# **Step 2.2 : Division of Data into Training and Test sets & Training of the Model**

```python
data = indian.tagged_sents(tagged_set)
train_data = data[:train_rows] 
test_data = data[test_rows:]

#Invoking tagger
pos_tagger = tnt.TnT()
pos_tagger.train(train_data)
pos_tagger.evaluate(test_data)
```

# **Step 2.3 :Performing POS Tagging on Hindi Dataset**

```python
sentence_to_be_tagged = "शुरुआत में क्रिकेट भेड़ के चारागाह या इसके किनारे खेला जाता था."

tokenized = nltk.word_tokenize(sentence_to_be_tagged)


print(pos_tagger.tag(tokenized))
```

# **Step 2.4: Performing POS Tagging with HMM**

```python
# Gettting all labels of the data
labels_set = set()
print("Adding labels from test sentences")
for test_sentence in test_data:
    for _,label in test_sentence:
        labels_set.add(label) 
print("Adding labels from training sentences")
for train_sentence in train_data:
    for _,label in train_sentence:
        labels_set.add(label) 
labels_set.add('Unk')
print("All labels:",labels_set)
```

```python
#Getting tags of the predicted sentances
pred_sents = pos_tagger.tag_sents([[word for word,_ in test_sent] for test_sent in test_data])
```

```python
# computing gold(true) label and predicted labels
gold_labels = [[glabel for _,glabel in test_sent] for test_sent in test_data]
pred_labels = [[plabel for _,plabel in pred_sent] for pred_sent in pred_sents]
```

```python
import numpy as np


# label to index lookup dictionary
labels_idxs = dict([labelidx for labelidx in zip(list(labels_set),range(0, len(labels_set)))])
# index to label lookup dictionary
idxs_labels = dict([(idx,label) for (label,idx) in labels_idxs.items()])

# confusion matrix for our counts
cm = np.zeros([len(labels_idxs),len(labels_idxs)],dtype=np.int)
# add up the prediction and gold context counts
for gold,pred in zip(gold_labels,pred_labels):
    for i in range(0,len(gold)):
        # get labels, look up indexes, add count to position
        cm[labels_idxs[gold[i]]][labels_idxs[pred[i]]] += 1
```

```python
import sys

# function to pretty print the confusion matrix
def pprintcm():
    maxlen = max(len(label) for label in labels_set)+1
    strformat = "{0:<"+str(maxlen)+"}"

    sys.stdout.write(strformat.format(''))
    for i in range(0,len(idxs_labels)):
        sys.stdout.write(strformat.format(idxs_labels[i]))
    print()
    idx=0
    for i in range(0,len(idxs_labels)):
        for j in range(0,len(idxs_labels)):
            if j == 0:
                sys.stdout.write(strformat.format(idxs_labels[idx]))
                sys.stdout.write(strformat.format(cm[i,j]))
                idx+=1
            else:
                sys.stdout.write(strformat.format(cm[i,j]))
        print()

# pretty print confusion matrix
pprintcm()
```
